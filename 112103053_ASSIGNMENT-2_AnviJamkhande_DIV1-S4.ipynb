```python
'''Define a class to represent a bank account. Include the following details like name of 
the depositor, account number, type of account, balance amount in the account. Write 
methods to assign initial values, to deposit an amount, withdraw an amount after 
checking the balance, to display name, account number, account type and balance.'''

import random
```

```python
# Creating a class for the bank account

class BankAcc :
    
    def __init__(self, bank_name, name, account_type, account_number = random.randint(100000,999999), balance = 0) : 
        self.bank_name = bank_name
        self.name = name
        self.account_number = account_number
        self.balance = 0
        self.account_type = account_type
        
    def withdraw(self, amount) :
        if (amount < self.balance) :
            print(f"Safely withdrawn Rs. {amount}")
            self.balance -= amount
        else :
            print("Insufficient balance!")
        print(f"Current balance: Rs. {self.balance}")
    
    def deposit(self, amount) :
        print(f"Successfully deposited Rs. {amount}")
        self.balance += amount
        print(f"Current balance: Rs. {self.balance}")
        
    def display_details(self) :
        print(f"Name of Bank: {self.bank_name}")
        print(f"Name of account owner: {self.name}")
        print(f"Account Number: {self.account_number}")
        print(f"Current bank balance: Rs. {self.balance}")
        print("Account type: ",end="")
        if (self.account_type == 1) :
            print("Savings Deposit")
        else :
            print("Fixed Deposit")
        
```

```python
# Creating an Account (Menu-driven) :

# Bank accounts will be stored in a list, an index will be pointing to the recent bank account

account_storage = []
l = []
choice = 1

while(choice) :
    
    print("Enter the appropriate digit for the desired output")
    print("1 - Create a bank account")
    print("2 - Withdraw money")
    print("3 - Deposit money")
    print("4 - Display bank details")
    print("0 - Quit")
    
    choice = int(input("Enter option: "))
    
    # Create Bank Account
    if (choice == 1) :
        bank_name = input("Enter name of the bank: ")
        name = input("Enter name of the account holder: ")
        print("Enter 1 for creating an FD account")
        print("Enter 2 for creating an SD account")
        account_type = input("Enter option: ")
        x = BankAcc(bank_name, name, account_type)
        l.append(x)
        
    # Withdraw money
    elif (choice == 2) :
        amount = int(input("Enter amount to withdraw: "))
        l[-1].withdraw(amount)
        
    # Deposit money    
    elif (choice == 3) :
        amount = int(input("Enter amount to deposit: "))
        l[-1].deposit(amount)
        
    # Display details 
    elif (choice == 4) :
        l[-1].display_details()
        
    # To filter out wrong input
    elif (choice != 0) :
        print("Wrong input, try again!")
        
print("Program has finished, thank you!")
```
