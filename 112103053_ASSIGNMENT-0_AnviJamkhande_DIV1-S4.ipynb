```python
#  Design and implement a class representing a rectangle and operations like area, perimeter, change dimensions, report dimensions, on it. Write code to test your classes.
```

```python
# rectangle:
class Rectangle: 
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
    def area(self):
        return self.length*self.breadth
    def perimeter(self):
        return 2*(self.length + self.breadth)
    def changeDimensions(self, new_length, new_breadth):
        self.length = new_length
        self.breadth = new_breadth
    def display(self):
        print(f"Length of the reactangle is: {self.length}")
        print(f"Breadth of the reactangle is: {self.breadth}")
```

```python
# Menu driven program

# Rectangles are stored in a list, the latest created rectangle is accessed by the user

choice = 1
l = []

while(choice) :
    
    print("1 - Create a rectangle")
    print("2 - Display length and breadth of rectangle")
    print("3 - Display area of the rectangle")
    print("4 - Display perimeter of the rectangle")
    print("5 - Modify dimensions")
    print("0 - Quit")
    print("\n")
    choice = int(input("Enter option: "))
    
    # Create a rectangle
    if (choice == 1) :
        x = Rectangle(4,2)
        l.append(x)
        
    # Display length and breadth 
    elif (choice == 2) :
        l[-1].display()
        
    # Display area 
    elif (choice == 3) :
        print(f"Area of rectangle: {l[-1].area()}")
    
    # Display perimeter 
    elif (choice == 4) :
        print(f"Perimeter of rectangle: {l[-1].perimeter()}")
        
    # Modify dimensions
    elif (choice == 5) :
        changed_length = int(input("Enter modified length: "))
        changed_breadth = int(input("Enter modified breadth: "))
        l[-1].changeDimensions(changed_length, changed_breadth)
        
    # Incorrect input
    elif (choice != 0) :
        print("Invalid input, try again!")        
```
